```python
# IMPORT LIBRARIES
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
theme.head()
```

```python
df.head()
```

```python
theme.head()
```

```python
# WE WANT TO MERGE THEME HEAD AND DF.HEAD
entire_data = df.merge(theme, left_on='parent_theme', right_on = 'name')
entire_data = entire_data.drop(columns='name_y')
entire_data.head()
```

```python
entire_data[entire_data['set_num'].isnull()].shape
```

```python
# TASK 1 : What percentage of all licensed sets ever released were Star Wars Themed?
licensed = entire_data[entire_data['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
# filter Star Wars 
star_wars = licensed[licensed["parent_theme"] == 'Star Wars']
# find how many 
the_force = int(star_wars.shape[0]/licensed.shape[0]*100)
print(the_force)
```
