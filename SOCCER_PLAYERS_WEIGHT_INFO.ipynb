```python
# IMPORT LIBRARIRES 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# READ DATA 
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)


```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 120].count()[0]
light_medium = fifa[(fifa.Weight >= 120) & (fifa.Weight < 135)].count()[0]
medium = fifa[(fifa.Weight >= 130) & (fifa.Weight < 160)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['Light Players', 'Light Medium Players', 'Medium Players', 'Medium Heavy Players', 'Heavy Players']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
